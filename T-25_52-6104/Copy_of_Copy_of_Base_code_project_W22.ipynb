##Given Helper Functions##

```python
def get_size(image):
  return image.size
```

```python
def get_pixel_value(img,col,row):
  return img.getpixel(col,row)
```

```python
def quantization(image, n):
  return image.quantize(n)
```

```python
def get_unique_values(image):
  return np.unique(image), len(np.unique(image))
```

```python
def extract_row_color(arr,row,color):
  filtered = filter(lambda item: ((item[0] == row) and (item[-1] == color)), arr)
  return list(filtered)
```

```python
def merge_row_color(filtered, row, color):
  remove_row_color = [item[1:-1] for item in filtered]
  merge_runs = (row,)
  for item in remove_row_color:
    merge_runs = merge_runs + item
  otp = merge_runs + (color,)
  return otp
```

##Task 2##
**Task 2: Compute the runs per each row for the image**

```python
def compute_runs(image, unique_values):
  runs_code = []
  # ......start code......
  for row in range(image.height):
    colomn = 0
    current_color = image.getpixel((0,row))
    for column in range(1, image.width):
        pixel_color =  image.getpixel((column,row));
        if pixel_color != current_color:
          runs_code.append((row, colomn, column-1, current_color))
          colomn = column
          current_color = pixel_color
    runs_code.append((row, colomn, image.width-1, current_color))
  return runs_code
  # ......end code......
  
```

##Task 3##
**Compute the Run-Length Encoding of an Image**

```python
def compute_RLE(image, unique_values):
    result = []
    run_code = []
    run_code = compute_runs(image, unique_values)
    # ......start code......
    for i in range (image.height):
      for j in range (len(unique_values)):
        temp= extract_row_color(run_code,i,unique_values[j])
        
        if len(temp) != 0:
          temp2 =merge_row_color(temp,i,unique_values[j])
          result.append(temp2)

    # ......end code......
    return result
```
